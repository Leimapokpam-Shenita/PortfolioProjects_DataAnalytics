```python
import os

def send_mail():
    EMAIL = os.getenv("EMAIL_USER")
    PASSWORD = os.getenv("EMAIL_PASS")

    server = smtplib.SMTP_SSL('smtp.gmail.com', 465)
    server.ehlo()
    server.login(EMAIL, PASSWORD)

    subject = "The i-phone you want is below 50000! Now is your chance to buy!"
    body = (
        "This is the moment we have been waiting for. "
        "Now is your chance to pick up the mobile of your dreams. "
        "Don't mess it up!\n\n"
        "Link here: https://www.amazon.in/Apple-iPhone-15-128-GB/dp/B0CHX2F5QT/"
    )

    msg = f"Subject: {subject}\n\n{body}"

    server.sendmail(
        EMAIL,
        EMAIL,
        msg
    )

    server.quit()
```
